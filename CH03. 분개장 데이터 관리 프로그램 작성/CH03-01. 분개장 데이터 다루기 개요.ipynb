# [ 3-1. 분개장 데이터 다루기 개요 ]

## 1. 분개장 데이터 검토
* 일반적으로 회계 프로그램에서 거래처원장, 계정별원장을 출력하여 회계 검토 수행

* 분개장 데이터 전체를 받아서 파이썬에 로딩한 후 계정별, 거래처별, 날짜별 분석 가능

    - 일반적으로 많이 사용하는 거래처원장, 계정별원장의 형태를 커스터마이징 하여 출력 가능
    
    - 거래처와 계정 구분을 복합적으로 엮어서 출력 가능
